## Escribiendo en un fichero existente

```python
fichero = open('Texto1.txt','a')  #extiendo el archivo, agrega escritura al final del texto en el archivo si lo hubiera (no sobreescribe)
fichero.write('\n continuando con ficheros en Python')
fichero.close()
fichero = open('Texto1.txt','r')  
text1 = fichero.read()
fichero.close()
print(text1)
```

## Moviendo el puntero

```python
fichero = open('Texto1.txt','r')
fichero.seek(3) # mueve el fichero al inicio posición 0
f1 = fichero.read(10) # leer las siguientes 5 bytes
fichero.seek(8) # mueve el fichero al cuarto byte
f2 = fichero.read(3) # leer las siguientes 5 bytes
print(f1)
print(f2)
print(fichero.tell()) #muestra posición del puntero
fichero.close()
```

## Escribiendo un archivo csv

```python
import csv
data = [["Raul", "19", "20181566"], ["Joel", "18", "20180101"], ["Carla", "19", "20181520"]]
with open('estudiantes.csv', 'w') as csvfile:
        #csvfile =open('estudiantes.csv', 'w')
    writer = csv.writer(csvfile, delimiter=',')
    writer.writerows(data) 
```
